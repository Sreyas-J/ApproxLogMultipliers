```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from mitchell.mitchell import approxMultiply
from ILM.ILMpy import logmult
```

```python
def plot(design,mul):
    error = []
    a_vals = []
    b_vals = []

    # Loop over values of a and b
    for a in range(256):
        for b in range(256):
            a_vals.append(a)
            b_vals.append(b)
            error.append(mul(a, b) - a * b)

    a_vals = np.array(a_vals)
    b_vals = np.array(b_vals)
    error = np.array(error)

    A = np.arange(256)
    B = np.arange(256)
    A, B = np.meshgrid(A, B)
    Error = error.reshape(256, 256)

        # Plot the 3D surface with grayscale colormap.
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(A, B, Error, cmap='gray', edgecolor='none')
    ax.set_xlabel('a')
    ax.set_ylabel('b')
    ax.set_zlabel('Error')
    ax.set_title(f"{design}'s Error")
    plt.show()

    print(f"{design}'s Maximum Error: ",np.min(error))
```

```python
plot("Mitchell",approxMultiply)
```

```python
plot("ILM",logmult)
```
